# M3 - Updated speedup-table

### General reflections:

- Based on the results seens with just Dask local the procedure seems to add way more overhead than simply using Parallel computing, 
- The LIF was significantly higher( 4.65 vs 0.75). The additional overhead of using Dask seems to negate the positive impact of the parallel/distributed computing. Only gaining a 1.4x speedup vs baseline. The speed was there, but due to the overhead and faster runtime, parallel computing is to be preferred, atleast at this problem size. At higher sizes the Dask approach could prove more beneficial.
